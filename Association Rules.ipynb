# BOOK

```python
import numpy as np
import pandas as pd 
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book = pd.read_csv('C://Users//Lenovo//Downloads//Assignment//Association Rules//book.csv')
book.head(5)
```

```python
frequent_books = apriori(book,min_support = 0.1,use_colnames=True)
frequent_books
```

```python
rules_books = association_rules(frequent_books,metric = 'lift',min_threshold = 0.5)
rules_books
```

```python
l = []
for i in range(1,100):
    if i%2==0:
        l.append(i)
```

```python
rules_books = rules_books.drop(l,axis=0)
```

```python
rules_books
```

```python
rules_books.sort_values('lift',ascending = False)[0:20]
```
